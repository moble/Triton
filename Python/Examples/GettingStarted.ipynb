## Getting started with *Project Triton* via python

First, we'll assume that you have compiled *Project Triton*, and everything has worked properly.Next, you need to make sure that your python path is set up properly, so that it can find the module.  You should add something like the following to your `~/.pythonrc` file:    import sys    sys.path.insert(1, '/path/to/project/Triton/Python')(Obviously, adjusting the path appropriately for your system.)Now, to see if that worked, try to evaluate the following by placing your cursor on the line, and hitting Shift-Enter.

```python
import PyGW
```

Assuming that your path did get set up properly, this should just quietly proceed, with no output.This automatically imports the basic module needed to interact with the *Project Triton* `c++` code.The fundamental object in this module is the `Waveform` object.  We can create an instance of this object by importing from disk.  This `git` repository comes with some basic test data in `Triton/Python/Examples/TestData`.  Assuming that you haven't changed directory in this ipython notebook session, you can get the data by doing

```python
W = PyGW.Waveform('TestData/rMPsi4_ExtrapolatedN-1.dat', 'MagArg')
```

This creates an object `W` by calling the `Waveform` constructor from the `PyGW` module.  The first argument is the data file (either a full path or a path relative to your pwd); the second argument is the format of the data -- either `'MagArg'` or `'ReIm'`.For now, the main use of the `Waveform` object is the set of methods you can use on it.  To see the methods available, just put your cursor after `W.` below, and hit Tab.

```python
W.
```

As you can see, there's quite a list.  Unfortunately, docstrings are not yet implemented for all these member variables, and even the parameter lists do not appear, due to quirks of SWIG.  For now, thebest way to find out how to use these member variables is to look at `Triton/Objects/Waveform.hpp`.One important caveat is that almost all member functions operate in place -- rather than returning a copy of the original `Waveform` object with the changes made, the original is changed and returned.Perhaps the most useful member functions to start out with are the getty functions.  For example, every`Waveform` object contains its history.  So we can get this and print it out with

```python
print(W.HistoryStr())
```

This shows us the git revision of *Project Triton* as currently compiled, the working directory, the hostname, and the command used to create the `Waveform`.  If you do anything more to `W`, it will be recorded, and printed out when the object is output.`Waveform` objects also contain the actual data, of course -- time and phase information, for example.There is a member function called `T()` and another called `Omega2m2()` [which gives the frequency of the (2,2) mode].  So we can plot these quantities:

```python
plot(W.T(), W.Omega2m2())
```

Now, because this is just a very short test data set, we only see the very earliest part of this $\sim 4000 M$ simulation, so we don't see the frequency sweep, and there's a lot of early junk.We can get a better look by zooming in.

```python
plot(W.T(), W.Omega2m2())xlim(100, 400)ylim(0.05, 0.07)
```
