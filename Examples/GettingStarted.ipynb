# Getting started with *Project Triton_IS_FOR_OLD_DATA* via python

## Introduction to *Project Triton_IS_FOR_OLD_DATA*

*Project Triton_IS_FOR_OLD_DATA* is the collection of code created by Mike Boyle for the purpose of analyzing and 
manipulating gravitational waveforms.  Some examples of uses built into the code are

* Extrapolation
* Convergence error estimation
* Construction of PN waveforms
* Comparison between PN and NR
* Hybridization of PN and NR
* Scaling to physical mass and distance
* Conversion to the frequency domain (trickier than it sounds)
* Evaluation of mismatches (*much* trickier than it sounds)

The code is designed to make these jobs as simple, automatic, and intuitive as possible.  From the 
user's perspective, the most important part of *Project Triton_IS_FOR_OLD_DATA* is the `Waveform` object.
This object encapsulates all the main capabilities of the code, and provides a handy interface.

For example, suppose that you have two `Waveform` objects, `W1` and `W2`, and you wish to align them 
by minimizing the squared phase difference over a time range $t = (100,1000)$.  You would issue the 
command

    W1.AlignTo(W2, 100, 1000)

This adjusts the time and phase data of `W1`, while leaving `W2` fixed.  In particular, it is important 
to note that almost all methods of the `Waveform` class operate on the object in place; rather than 
returning an altered copy of the original, the methods just alter the original.

We will see later how to construct `Waveform` objects, and see all the various functions we can use with 
them.

## Obtaining and building *Project Triton_IS_FOR_OLD_DATA*

Dependencies:

    * Other compilers would require minimal changes to the `Makefile`s

* SWIG 2.0 or greater (this creates the python bindings to c++ functions)

Note that the friendliest environment for using python is provided by `ipython --pylab`.  That is the
environment this package is developed in, so getting the most recent version should be sufficient.

## Important caveats for using `PyGW_IS_FOR_OLD_DATA`
### Python assignment makes shallow copies
One thing that may be dangerous for c++ programmers trying to use python is the assignment "operator".
In fact, python has no such thing for fundamental objects.  Rather, it has an assignment *statement*, which
does not actually copy objects.  For example, if you have a `Waveform` object `W1` and want to construct a 
copy of it, you can't just do

    W2 = W1

In this case, any change you make to `W1` will happen to `W2` (and vice versa) because they are both just
references to the underlying object.  Instead, what you want to do is explicitly construct a new `Waveform`
object which contains a copy of the original:

    W2 = PyGW_IS_FOR_OLD_DATA.Waveform(W1)

This calls the c++ copy constructor.  Do ***not*** try to be clever and use the 
[copy module](http://docs.python.org/library/copy.html) for deep copies.  This is not guaranteed to work.

Note that this is not an issue with SWIG or *Project Triton_IS_FOR_OLD_DATA*; this is a fundamental issue to keep in mind
whenever you use python.

### Methods typically operate in place
As mentioned previously, methods for the `Waveform` object (and others we will meet) prefer to operate in
place, rather than returning a new, altered copy of the object.  This is because `Waveform` objects can 
become extremely large (many gigabytes) for some applications, so operating in place can save 
large amounts of memory and improve speed.

That said, there are several methods that fundamentally alter the type of data returned, or cannot benefit
from , so they do return
a new `Waveform` object.  At the time of this writing, the only such methods are

* `operator[]`, which returns just one mode of the `Waveform`
* `operator/`, which returns the difference between two `Waveform`s
* `HybridizeWith` and `HybridizeWith_F`, which hybridize pairs of `Waveform`s

## Member data and access methods of the `Waveform` object

The member data contained in each `Waveform` object are:

    std::stringstream history;
    unsigned int typeIndex;
    std::string timeScale;
    std::vector<double> t;
    std::vector<double> r;
    std::vector<WaveformUtilities::Quaternion> frame;
    WaveformUtilities::Matrix<int> lm;
    WaveformUtilities::Matrix<double> mag;
    WaveformUtilities::Matrix<double> arg;

These are each described in the following subsections.

### `history`
This contains the history of the object, beginning with the construction during the present 
session, followed by any previous history, followed by what has happened  to the `Waveform` in this
session since its construction.

```python
HistoryString = W.HistoryStr()
print(HistoryString[0:HistoryString.index('Begin Previous History')+22]+'\n...')
```

This shows us the git revision of *Project Triton_IS_FOR_OLD_DATA* as currently compiled, the working directory, the 
hostname, and the command used to create the `Waveform`.  The section containing the previous history
is not shown, as it is quite long.  If you do anything more to `W`, it will be recorded, and printed 
out when the object is output.

### `typeIndex`

This is a digit corresponding to one of the types of `Waveform`:
<ol type="1">
<li> rMPsi4 </li>
<li> rhdot </li>
<li> rhOverM </li>
<li> rPsi4 </li>
<li> rhdot </li>
<li> rh </li>
<li> MPsi4 </li>
<li> hdot </li>
<li> hOverM </li>
<li> Psi4 </li>
<li> hdot </li>
<li> h </li>
</ol>
The object tracks which type it is, and uses this information when being rescaled, transformed, etc.  
The index itself may be accessed from outside the class with the member function `TypeIndex()`, while
the corresponding string is returned by `Type()`.

### `timeScale`

This is simply a string describing the time scale.  Typically, this would be `(t-r*)/M`.  
The time scale string is accessed from outside the class with the member function `TimeScale()`.

### `t`

This is the vector of times.  All time-dependent data are assumed to have the same length as this
vector.  The length of this vector is accessed from outside the class by the member function `NTimes()`.

### `r`

The vector of the radii at which the `Waveform` is measured.  This vector may have length 0, 1, or 
`t.size()`.  If length 1, it is assumed to be constant throughout the data set.

### `frame`

This is a vector of `Quaternion` objects describing the orientation of the frame relative to the stationary
frame.  More precisely, if the standard frame is described by basis vectors $(\vec{x}, \vec{y}, \vec{z})$, 
then the frame in which the waveform is measured has basis vectors $(q\vec{x}\bar{q}, q\vec{y}\bar{q}, 
q\vec{z}\bar{q})$, where $q$ is the quaternion at a particular instant.

Again, this vector may have length 0, 1, or `t.size()`, as with `r`.
