# Example contents for ipython notebooks

The notebook interface to ipython may be loaded with a command line like

  `ipython notebook --pylab inline`

The `--pylab` option loads pylab, of course.  The `inline` option allows graphics to be displayed in the notebook, rather than opening an external python window.  For example:

```python
plot([i**3 for i in range(20)], 'b-o')
xlabel(r'$(t-r_\ast)/M$')
ylabel(r'$\|\Psi_4\|$')
show()
```

In Markdown cells, LaTeX may be entered as usual, as in $\Psi_4$.  Note that MathJax is apparently used, so you can expect the syntax to match its quirks and capabilities.

\begin{equation}
  \Psi_4
\end{equation}

[Standard markdown](http://en.wikipedia.org/wiki/Markdown) and <em>html</em> are also allowed.

Use single asterisks or single underlines for *emphasis* and _italics_.

Use double asterisks or double underlines for **bold** (though be careful of __bugs__).

Use backticks for `tt font`.  Alternatively, code indented by four spaces is syntax highlighted:

    int main() {
        return 0;
    }

Use square brackets followed by parentheses to enclose a web site name and its url as in [black-holes](http://www.black-holes.org).
Note, however, that footnotes don't work.  Fair enough.

Horizontal rules are created by placing three or more asterisks, hyphens, or underscores (with or without spaces) on a separate line, with empty lines on either side.

Itemized lists are possible

    * indent consecutive levels by four spaces
        * and begin each item with `*` or `-`
            * and continue to four levels

    1. like this: `1.`

    1. Instead, the numbers
        1. and the sets from which they are chosen
            1. are chosen automatically
                1. but only to the fourth/fifth level of indentation
