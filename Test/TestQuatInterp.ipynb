```python
import PyGW.plot_on_sphere
import numpy as npy
import matplotlib.pyplot as plt

mpl.rcParams['figure.figsize'] = 9,6  # Set default figure size (in inches)
%config InlineBackend.figure_format='svg'

vartheta, varphi = PyGW.plot_on_sphere.sphere_coordinates()
x,y,z = PyGW.plot_on_sphere.sphere_points(vartheta, varphi)
QuatInterpDat = npy.genfromtxt('TestQuatInterp.dat')
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Surface = ax.plot_surface(0.95*x, 0.95*y, 0.95*z,
                          rstride=10, cstride=10, linewidth=0.5, antialiased=False)
Dots1 = ax.plot(QuatInterpDat[:5,1], QuatInterpDat[:5,2], QuatInterpDat[:5,3], 'go', ms=10)
Dots2 = ax.plot(QuatInterpDat[5:,1], QuatInterpDat[5:,2], QuatInterpDat[5:,3], 'ro')
```

The above should show a sphere with various red dots, which depict the history of the $z$ axis as they are rotated by the various quaternions output by the `TestQuatInterp` routine.

The basic upshot is that the curve should look pretty smooth, and should go through the locations corresponding to the `qIn` variable in `TestQuatInterp.cpp`, which should correspond to the green dots.
